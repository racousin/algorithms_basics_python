```python
from scipy.stats import *
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
np.random.seed(seed=233423)
```

```python
n = 10000
```

```python
unif = uniform(loc=0, scale=1)
```

```python
u = unif.rvs(n)
```

# eponential

Density: $f(x) = \lambda exp(-\lambda x)$ with $x, \lambda \in \mathbb{R}^+$

Repartition function: $F(x) = 1 - exp(-\lambda x)$

Inverse repartition function: $F^{-1}(u) = \log (\frac{1}{1 - u}) / \lambda$

```python
lamb = np.pi
```

```python
def inv_exp(u, lamb):
    return np.log(1/u) / lamb
```

```python
X = inv_exp(u,lamb)
xs = np.linspace(X.min(), X.max()/2,100)
pdf = expon(loc=0, scale = 1/lamb).pdf(xs)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.distplot(X,bins=None,hist=0, label="approx pdf")
ax.set_xlim(X.min(),X.max()/2)
ax.plot(xs,pdf,'r-', label="exact pdf")
ax.legend()
```

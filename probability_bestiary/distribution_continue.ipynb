### Uniform
$X \sim \mathbb{U}[a,b]$

$f(x) = \frac{1}{b - a}\mathbb{1}_{[a,b]}(x)$

$F(x) = \frac{x - a}{b - a}\mathbb{1}_{[a,b]}(x) + \mathbb{1}_{]b,+\infty[}(x)$

$E[X] = \frac{a+b}{2}$

$V(X)$ = $\frac{(b-a)^2}{12}$ 

```python
dunif = lambda n, a, b: uniform(loc=a,scale=(b-a)).pdf(x) 
punif = lambda x, a, b: uniform(loc=a,scale=(b-a)).cdf(x)
runif = lambda x, a, b: uniform(loc=a,scale=(b-a)).rvs(n)
qunif = lambda q, a, b: uniform(loc=a,scale=(b-a)).ppf(q)
```

```python
x = np.linspace(-6, 6, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

a_ = np.arange(-5,5,2)
b_ = [a + np.random.randint(5) for _ in a_]
for i in range(len(a_)):
    ax[0].plot(x, dunif(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Probability density function
    ax[1].plot(x, punif(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Normal
$X \sim \mathbb{N}(m,\sigma^2)$

$f(x) = \frac{1}{\sqrt{2\pi \sigma^2}}\exp(-\frac{1}{2 \sigma^2}(x - m)^2)$

$F(x) = \frac{1}{2}[1 + erf(\frac{x - \mu}{\sigma\sqrt(2)})]$

$E[X] = m$

$V(X)$ = $\sigma^2$ 

```python
dnorm = lambda n, m, sig: norm(loc=m,scale=sig).pdf(x) 
pnorm = lambda x, m, sig: norm(loc=m,scale=sig).cdf(x)
rnorm = lambda x, m, sig: norm(loc=m,scale=sig).rvs(n)
qnorm = lambda q, m, sig: norm(loc=m,scale=sig).ppf(x)
```

```python
x = np.linspace(-10, 10, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

m_ = np.arange(-5,5,2)
sig2_ = [round(np.random.rand()*10 + 0.2,2) for _ in m_]
for i in range(len(m_)):
    ax[0].plot(x, dnorm(x, m_[i], np.sqrt(sig2_[i])), label=r'$m={},\sigma^2={}$'.format(m_[i],sig2_[i]))  # Probability density function
    ax[1].plot(x, pnorm(x, m_[i], np.sqrt(sig2_[i])), label=r'$m={},\sigma^2={}$'.format(m_[i],sig2_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Exponential
$X \sim \epsilon (\lambda)$

$f(x) = \lambda e^{-\lambda x}\mathbb{1}_{\mathbb{R}^+_*}(x)$

$F(x) = 1 - e^{-\lambda x}\mathbb{1}_{\mathbb{R}^+_*}(x)$

$E[X] = \frac{1}{\lambda}$

$V(X)$ = $\frac{1}{\lambda^2}$ 

```python
dexp = lambda n, lamb: expon(loc=0,scale=lamb).pdf(x) 
pexp = lambda x, lamb: expon(loc=0,scale=lamb).cdf(x)
rexp = lambda x, lamb: expon(loc=0,scale=lamb).rvs(n)
qexp = lambda q, lamb: expon(loc=0,scale=lamb).ppf(q)
```

```python
x = np.linspace(0, 10, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

lamb_ = np.arange(1,10,2)
for i in range(len(lamb_)):
    ax[0].plot(x, dexp(x, lamb_[i]), label=r'$lamb={}$'.format(lamb_[i]))  # Probability density function
    ax[1].plot(x, pexp(x, lamb_[i]), label=r'$lamb={}$'.format(lamb_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Gamma
$X \sim \mathbb{N}(a,b)$

$f(x) = \frac{b^a}{\Gamma(a)}x^{a - 1}e^{-bx}\mathbb{1}_{\mathbb{R}^+_*}(x)$

$F(x) = \frac{1}{\Gamma(a)}\gamma(a,bx)\mathbb{1}_{\mathbb{R}^+_*}(x)$

$E[X] = \frac{a}{b}$

$V(X)$ = $\frac{a}{b^2}$ 

```python
dgamma = lambda n, a, b: gamma(a=a,scale=b).pdf(x) 
pgamma = lambda x, a, b: gamma(a=a,scale=b).cdf(x)
rgamma = lambda x, a, b: gamma(a=a,scale=b).rvs(n)
qgamma = lambda q, a, b: gamma(a=a,scale=b).ppf(q)
```

```python
x = np.linspace(0, 20, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

a_ = np.arange(1,10,2)
b_ = [round(np.random.rand()*2 + 0.2,2) for _ in a_]
for i in range(len(a_)):
    ax[0].plot(x, dgamma(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Probability density function
    ax[1].plot(x, pgamma(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Beta
$X \sim Beta(a,b)$

$f(x) = x^{a - 1}(1 - x)^{b-1}\mathbb{1}_{[0,1]}(x)$

$F(x) = I_x(a,b)$

$E[X] = \frac{a}{a+b}$

$V(X)$ = $\frac{ab}{(a+b)^2(a+b+1)}$ 

```python
dbeta = lambda n, a, b: beta(a=a,b=b).pdf(x) 
pbeta = lambda x, a, b: beta(a=a,b=b).cdf(x)
rbeta = lambda x, a, b: beta(a=a,b=b).rvs(n)
qbeta = lambda q, a, b: beta(a=a,b=b).ppf(q)
```

```python
x = np.linspace(0, 1, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

a_ = [round(np.random.rand()*5 + 0.5,2) for i in range(5)]
b_ = [round(np.random.rand()*5 + 0.5,2) for i in range(5)]
for i in range(len(a_)):
    ax[0].plot(x, dbeta(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Probability density function
    ax[1].plot(x, pbeta(x, a_[i], b_[i]), label=r'$a={},b={}$'.format(a_[i],b_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### $\chi^2$
$X \sim \chi^2(k)$

$f(x) = \frac{1}{2^{k/2}\Gamma(k/2)x^{k/2 -1}e^{-x/2}}\mathbb{1}_{\mathbb{R}^+_*}(x)$

$F(x) = \frac{1}{\Gamma(k/2)}\gamma(k/2,x/2)\mathbb{1}_{\mathbb{R}^+_*}(x)$

$E[X] = k$

$V(X)$ = $2k$ 

```python
dchisq = lambda n, k: chi2(df=k).pdf(x) 
pchisq = lambda x, k: chi2(df=k).cdf(x)
rchisq = lambda x, k: chi2(df=k).rvs(n)
qchisq = lambda q, k: chi2(df=k).ppf(q)
```

```python
x = np.linspace(0, 10, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

k_ = np.arange(1,10,2)
for i in range(len(lamb_)):
    ax[0].plot(x, dchisq(x, k_[i]), label=r'$lamb={}$'.format(k_[i]))  # Probability density function
    ax[1].plot(x, pchisq(x, k_[i]), label=r'$lamb={}$'.format(k_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Student
$X \sim t(v)$

$X = \frac{Z}{\sqrt(V/v)}$, where $Z \sim \mathbb{N}(0,1)$, $V \sim \chi^2(v)$ and $Z$ and $V$ independent 

$E[X] = 0$ if $v\geq 1$

$V(X)$ = $\frac{v}{v-2}$ if $v > 2$, $\infty$ otherwise

```python
dt = lambda n, k: t(df=k).pdf(x) 
pt = lambda x, k: t(df=k).cdf(x)
rt = lambda x, k: t(df=k).rvs(n)
qt = lambda q, k: t(df=k).ppf(q)
```

```python
x = np.linspace(-5, 5, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

k_ = np.arange(1,10,2)
for i in range(len(lamb_)):
    ax[0].plot(x, dt(x, k_[i]), label=r'$v={}$'.format(k_[i]))  # Probability density function
    ax[1].plot(x, pt(x, k_[i]), label=r'$v={}$'.format(k_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Fisher
$X \sim \mathcal{F}(d_1,d_2)$

$X = \frac{S_1/d_1}{S_2/d_2}$, where $S_1 \sim \chi^2(d_1)$,  $ S_2 \sim \chi^2(d_2)$ and $S_1$ $S_2$ independant

$E[X] = \frac{d_2}{d_2-2}$, for $d_2 > 2$

$V(X)$ = $\frac{2d^2_2(d_1 + d_2 -2)}{d_1(d_2 - 2)^2(d_2-4)}$, for $d_2 > 4$

```python
df = lambda n, d1, d2: f(dfn=d1, dfd=d2).pdf(x) 
pf = lambda x, d1, d2: f(dfn=d1, dfd=d2).cdf(x)
rf = lambda x, d1, d2: f(dfn=d1, dfd=d2).rvs(n)
qf = lambda q, d1, d2: f(dfn=d1, dfd=d2).ppf(q)
```

```python
x = np.linspace(0.1, 5, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

df1_ = [np.ceil(np.random.rand() * 20) for i in range(5)]
df2_ = [np.ceil(np.random.rand() * 20) for i in range(5)]
for i in range(len(df1_)):
    ax[0].plot(x, df(x, df1_[i], df2_[i]), label=r'$d1={}, d2={}$'.format(df1_[i], df2_[i]))  # Probability density function
    ax[1].plot(x, pf(x, df1_[i], df2_[i]), label=r'$d1={}, d2={}$'.format(df1_[i], df2_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Cauchy
$X \sim Cauchy (o,a)$

$f(x) = \frac{1}{\pi a \left[1 + (\frac{x - o}{a})^2\right]}$

$F(x) = \frac{1}{\pi} \arctan (\frac{x-o}{a}) + \frac{1}{2}$

$E[X] = undefined$

$V(X) = undefined$ 

```python
dcauchy = lambda n, o, a: cauchy(loc=o,scale=a).pdf(x) 
pcauchy = lambda x, o, a: cauchy(loc=o,scale=a).cdf(x)
rcauchy = lambda x, o, a: cauchy(loc=o,scale=a).rvs(n)
qcauchy = lambda q, o, a: cauchy(loc=o,scale=a).ppf(q)
```

```python
x = np.linspace(-10, 10, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

o_ = [np.round((np.random.rand()-.5) * 5,2) for i in range(5)]
a_ = [np.round((np.random.rand()) * 5 + 0.01,2) for i in range(5)]
for i in range(len(o_)):
    ax[0].plot(x, dcauchy(x, o_[i], a_[i]), label=r'$o={}, a={}$'.format(o_[i], a_[i]))  # Probability density function
    ax[1].plot(x, pcauchy(x, o_[i], a_[i]), label=r'$o={}, a={}$'.format(o_[i], a_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

### Laplace
$X \sim Laplace(o,a)$

$f(x) = \frac{1}{2a}exp(-\frac{|x -o|}{a})$

$F(x) = \frac{1}{2}exp(-\frac{o-x}{a})$ if $x \leq o$, $1 - \frac{1}{2}exp(-\frac{o-x}{a})$ otherwise

$E[X] = o$

$V(X)$ = $2a^2$  

```python
dlaplace = lambda n, o, a: laplace(loc=o,scale=a).pdf(x) 
plaplace = lambda x, o, a: laplace(loc=o,scale=a).cdf(x)
rlaplace = lambda x, o, a: laplace(loc=o,scale=a).rvs(n)
qlaplace = lambda q, o, a: laplace(loc=o,scale=a).ppf(q)
```

```python
x = np.linspace(-10, 10, num=200)
fig, ax = plt.subplots(1,2, figsize=(20,10))

o_ = [np.round((np.random.rand()-.5) * 5,2) for i in range(5)]
a_ = [np.round((np.random.rand()) * 5 + 0.01,2) for i in range(5)]
for i in range(len(o_)):
    ax[0].plot(x, dlaplace(x, o_[i], a_[i]), label=r'$o={}, a={}$'.format(o_[i], a_[i]))  # Probability density function
    ax[1].plot(x, plaplace(x, o_[i], a_[i]), label=r'$o={}, a={}$'.format(o_[i], a_[i]))  # Cumulative density function
ax[0].set_title('PDF')
ax[1].set_title('CDF')
ax[0].legend(loc="upper left");
ax[1].legend(loc="lower right");
```

```python
def grad_hinge(a,b,x):
    if b * x.T.dot(a) > 1:
        return np.zeros(len(x))
    else:
        return -b * a

def grad(a,b,x,lam):
    tmp = 0
    for i in range(len(b)):
        tmp += grad_hinge(a[i,:],b[i],x)
    return lam * tmp/len(b) + x

def grad_desc(a,b,T,x,lam):
    its = np.zeros(shape=(T+1,a.shape[1]))
    its[0,:] = x
    for i,t in enumerate(range(T)):
        s = 2 / (t+1)
        x = x - s * grad(a,b,x,lam)
        its[i+1,:] = x
    return x, its
def hinge(a,b,x):
    threshold = x.T.dot(a) * b
    cost = 1 - threshold
    cost[threshold >= 1] = 0
    return mean(cost)
```
